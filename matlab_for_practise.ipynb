**MATPLOTLIB**

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([2,4,6],[12,15,17])
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([2,4,6],[12,15,17])
plt.title("Ankita's Line Plotting")
plt.xlabel("X---->")
plt.ylabel("Y---->")
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot")
x1=[2,4,6]
y1=[21,34,45]
x2=[3,8,7]
y2=[32,47,51]
plt.plot(x1,y1,label="first",linewidth=4)
plt.plot(x2,y2,label="second",linewidth=4)
plt.legend()
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.scatter([2,4,6,11,14],[12,15,17,4,19])
plt.title("Ankita's Scatter Plotting")
plt.xlabel("X---->")
plt.ylabel("Y---->")
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
x=np.random.normal(170,10,250)
plt.hist(x)
plt.title("Ankita's Histogram Plotting")
plt.xlabel("X---->")
plt.ylabel("Y---->")
plt.show()
```
